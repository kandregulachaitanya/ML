###  1. Apply Linear Regression to the given dataset without splitting dataset

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Reading data
data = pd.read_csv('https://raw.githubusercontent.com/SumanaBethala/Problems/master/Lifeexpectancy.csv')
# Collecting X and Y
X = data['fertility'].values
print(X)
Y = data['life'].values
print(Y)
```

```python
m = len(X)
print(m)
X = X.reshape((m,1))
Y = Y.reshape((m,1))
print(X.shape)
print(Y.shape)
```

```python
reg = LinearRegression()
# Data Fitting
reg=reg.fit(X,Y)
# Y Prediction
Y_pred = reg.predict(X)
Y_pred
```

### 2. Check the R squared score

```python
reg.score(X,Y)
```

### 3. Split the dataset into training and testing sets.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state=1)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape,sep='\n')
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
train_score = lr.score(x_train,y_train)
test_score = lr.score(x_test,y_test)
print("Linear Regression train score:",train_score)
print("LinearRegression test score: ",test_score)
```

```python
plt.scatter(x_train,y_train,color='blue')
plt.scatter(x_test,y_test,color='red')
plt.xlabel('fertility')
plt.ylabel('life')
plt.show()
```

### 4. Check the R squared score

```python
train_score = lr.score(x_train,y_train)
test_score = lr.score(x_test,y_test)
print("Linear Regression train score:",train_score)
print("LinearRegression test score: ",test_score)
```

### 5. Finally take any value for fertility and predict the life.

```python

x_gen = np.linspace(1,8,100)
y_lnr = lr.predict(x_gen.reshape(-1,1))

plt.scatter(np.array(x_train),y_train,color='blue')
plt.scatter(np.array(x_test),y_test,color='red')
plt.scatter(x_gen,y_lnr,s=4,color='orange',label='Linear Regression')
plt.legend()
plt.show()
```
